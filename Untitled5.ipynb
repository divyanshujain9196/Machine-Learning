```python
import numpy as np



x=np.array([1,3,5,7,9,11])

y=np.array([13,19,23,25,29,31])







sum_x=np.sum(x)

sum_x

sum_y=np.sum(y)
sum_y

mul_xy=np.multiply(x,y)
mul_xy

sum_xy=sum(mul_xy)

sum_xy

n = 6
x_x=np.multiply(x,x)
sum_x_x=np.sum(x_x)
sum_x_x


p=np.array([[n,sum_x],[sum_x,sum_x_x]] )
p

q = np.array([sum_y, sum_xy])

q

z = np.linalg.solve(p, q)
z

y_ = np.array(z[0:1]+z[1:2]*x)

y_

avg = np.average(y)
# avg=int(avg)
avg

ss= np.subtract(avg,y_)
s=ss*ss
s

ssr=np.sum(s)

ssr

se=np.subtract(y,y_)
se*se

sse=np.sum(se*se)

sse

st=np.subtract(y,avg)
st*st

sst=np.sum((st*st))

sst

r_r=ssr/sst
r_r



percentage=r_r*100;

percentage

import matplotlib.pyplot as plt

plt.plot(x, y)
plt.show()

plt.scatter(x,y)
plt.plot(x,y_)
plt.xlabel("x")
plt.ylabel("y")

plt.plot(x,y_)
```
